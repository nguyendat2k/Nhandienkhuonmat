```python
#  tạo dataSet dùng để nhận diện
import numpy as np
import os
import sqlite3
import cv2
#--------------------------------------------------------------------
# CODE NHAP DU LIEU HINH ANH VA DAT TEN KET NOI CO SO DU LIEU
#insert/update data to sqlite
def insertOrUpdate(id, name):
    #connecting to the db
    conn =sqlite3.connect("Face.db")
    #check if id already exists
    query = "SELECT * FROM Face WHERE ID="+str(id)
    #returning the data in rows
    cursor = conn.execute(query)
    isRecordExist=0
    for row in cursor:
        isRecordExist=1
    if isRecordExist==0:
         query="INSERT INTO Face(ID, Name) VALUES("+str(id)+",'"+str(name)+"')"     
    else:
          query="UPDATE Face SET Name='"+str(name)+" WHERE ID="+str(id)
    conn.execute(query)
    conn.commit()
    conn.close()

# khai báo 1 lớp nhận diện khuôn mặt lưu vào 1 biến
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

# mở camera. Ở đây sài camera trên lap nên tham số là 0 .
cap = cv2.VideoCapture(0)

id = input("Enter user id:" )
name = input("Enter name: ")
insertOrUpdate(id, name)
sample_number = 0
while True:
    
    # dịch chuyển cái ô hình vuông cho vào mặt mình
    ret, img = cap.read()
    
    # Tạo một bức ảnh xám từ bức ảnh gốc
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
# Sử dụng phương thức detectMultiScale để phát hiện khuôn mặt trong bức ảnh xám
    # gray là nguồn/bức ảnh xám.
    # scaleFactor  = 1.3 :độ scale sau mỗi lần quét, tính theo 0.01 = 1%. Nếu như để scaleFactor = 1 thì tấm ảnh sẽ giữ nguyên
    # minNeighbors = 5 minNeighbors: scale và quét ảnh cho đến khi không thể scale được nữa thì lúc này sẽ xuất hiện những khung ảnh trùng nhau, số lần trùng nhau chính là tham số minNeighbors để quyết định cho việc có chọn khung ảnh này là khuôn mặt hay không.
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        #incrementing sample number 
        sample_number += 1

        if not os.path.exists("data_face"):
            os.makedirs("data_face")

#lưu khuôn mặt đã chụp vào thư mục 
        cv2.imwrite("data_face/User."+str(id)+"."+str(sample_number)+".jpg",  img[y:y+h,x:x+w])
    
# Vẽ một hình tứ giác xung quanh những khuôn mặt phát hiện được. Vẽ trên ảnh màu.
        cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
# # Hiển thị kết quả ra màn hình
    cv2.imshow("img", img)
    
    #Kết thúc chương trình
    cv2.waitKey(1);
    # break if the sample number is morethan 100
    if(sample_number>100):
        cap.release()
        cv2.destroyAllWindows()
        break;

```

```python
#Tạo lớp traning dữ liệu
from PIL import Image
```

```python
recognizer = cv2.face.LBPHFaceRecognizer_create()
path = 'data_face'
```

```python
def getImagesWithID(path):
   # Load đường dẫn và file trong thư mục dataSet
    imagePaths=[os.path.join(path, f) for f in os.listdir(path)]
    faces=[]
    IDs=[]
    for imagePath in imagePaths:
        #Convert ảnh và add vào mảng faces cùng với ID
        faceImg = Image.open(imagePath).convert('L')
        faceNp = np.array(faceImg, 'uint8')
        
       # tách ID từ tên file ảnh
        ID=int(os.path.split(imagePath)[-1].split('.')[1])
        
        faces.append(faceNp)
        IDs.append(ID)
        # Hiển thị kết quả ra màn hình
        cv2.imshow('training', faceNp)
        cv2.waitKey(10)
    return np.array(IDs), faces

Ids, faces = getImagesWithID(path)
recognizer.train(faces, Ids)

if not os.path.exists('trainer'):
    os.makedirs('trainer')
    
#Traning và lưu vào file
recognizer.save('huanluyen/huanluyen.yml')
cv2.destroyAllWindows()
```

```python
#.Lớp detector nhận diện khuôn mặt đã traning
import pickle, sqlite3

face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read("huanluyen/huanluyen.yml")

#Lấy thông tin user từ Database
def getProfile(Id):
    conn=sqlite3.connect("Face.db")
    query="SELECT * FROM Face WHERE ID="+str(Id)
    cursor=conn.execute(query)
    profile=None
    for row in cursor:
        profile=row
    conn.close()
    return profile

# mở camera. Ở đây sài camera trên lap nên tham số là 0 .
cap = cv2.VideoCapture(0)
font = cv2.FONT_HERSHEY_COMPLEX
while True:
    
    # dịch chuyển cái ô hình vuông cho vào mặt mình
    ret, img = cap.read()
    
    # Tạo một bức ảnh xám từ bức ảnh gốc
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    
   # Hiển thị thông tin User bằng text
    for (x,y,w,h) in faces:
        
    # Vẽ một hình tứ giác xung quanh những khuôn mặt phát hiện được. Vẽ trên ảnh màu.
        cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]

        nbr_predicted, conf = recognizer.predict(gray[y:y+h, x:x+w])
        if conf < 70:
            profile=getProfile(nbr_predicted)
            if profile != None:
                cv2.putText(img, ""+str(profile[1]), (x+10, y), font, 1, (0,255,0), 1);
        else:
            cv2.putText(img, "Unknown", (x, y + h + 30), font, 0.4, (0, 0, 255), 1);

# Hiển thị kết quả ra màn hình
    cv2.imshow("img", img)
# nhấn q để thoát chương trình
    if(cv2.waitKey(1) == ord("q")):
        break
cap.release()
cv2.destroyAllWindows()

```
